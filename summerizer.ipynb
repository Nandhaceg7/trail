```python
from transformers import pipeline

# Load summarizer
summarizer = pipeline("summarization")

# Example input
article = """
India's space agency ISRO has successfully launched the Chandrayaan-3 mission,
aimed at exploring the moon's south pole. The mission marks a significant
milestone in India's space journey.
"""

# Summarize
summary = summarizer(article, max_length=50, min_length=25, do_sample=False)
print("Summary:", summary[0]['summary_text'])
```
